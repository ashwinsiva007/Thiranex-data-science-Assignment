```python
import pandas as pd

df = pd.read_csv('Titanic-Dataset csv.csv')

# Check missing values
print(df.isnull().sum())

# Fill or drop missing values
df.fillna(df.mean(numeric_only=True), inplace=True)  # fill numbers with mean
# df.dropna(inplace=True)  # OR drop rows with missing values

# Remove duplicates
df.drop_duplicates(inplace=True)

print("Cleaned shape:", df.shape)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Basic stats
print(df.describe())

# Plot 1 - Age Distribution
sns.histplot(df['Age'], kde=True)
plt.title("Age Distribution")
plt.show()

# Plot 2 - Survival Count
df['Survived'].value_counts().plot(kind='bar', color=['red','green'])
plt.title("Survival Count (0=No, 1=Yes)")
plt.xlabel("Survived")
plt.show()

# Plot 3 - Correlation Heatmap
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
fig.suptitle("Titanic Dataset - Data Cleaning & Visualization Report", fontsize=16, fontweight='bold')

# Plot 1 - Age Distribution
sns.histplot(df['Age'], kde=True, ax=axes[0,0], color='skyblue')
axes[0,0].set_title('Age Distribution')

# Plot 2 - Survival Count
df['Survived'].value_counts().plot(kind='bar', ax=axes[0,1], color=['red','green'])
axes[0,1].set_title('Survival Count (0=No, 1=Yes)')

# Plot 3 - Passenger Class Count
df['Pclass'].value_counts().plot(kind='bar', ax=axes[0,2], color='orange')
axes[0,2].set_title('Passenger Class Distribution')

# Plot 4 - Fare Distribution
sns.boxplot(data=df, y='Fare', ax=axes[1,0], color='lightgreen')
axes[1,0].set_title('Fare Boxplot')

# Plot 5 - Survival by Gender
sns.countplot(data=df, x='Sex', hue='Survived', ax=axes[1,1])
axes[1,1].set_title('Survival by Gender')

# Plot 6 - Correlation Heatmap
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', ax=axes[1,2])
axes[1,2].set_title('Correlation Heatmap')

plt.tight_layout()
plt.savefig('titanic_dashboard.png', dpi=150, bbox_inches='tight')
plt.show()

print("✅ Dashboard saved as titanic_dashboard.png")
```
